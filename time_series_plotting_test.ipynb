```python
dates = [
         dt.datetime(2019, 5, 24),
         dt.datetime(2019, 5, 25),
         dt.datetime(2019, 5, 26),
         dt.datetime(2019, 5, 27),
         dt.datetime(2019, 5, 28),
         dt.datetime(2019, 5, 29),
         dt.datetime(2019, 5, 30),
]
```

```python
dates
```

```python
y = [0,1,3,4,6,5,7]
```

```python
def show_plot():
  plt.title('Bit Coin Prices')
  plt.xlabel('Date')
  plt.ylabel('Price')

  plt.gcf().autofmt_xdate()
  date_format = mpl_dates.DateFormatter('%b %d, %y')
  plt.gca().xaxis.set_major_formatter(date_format)

  plt.tight_layout()
  

  # plt.legend()

  plt.show()
```

```python
plt.plot(dates, y, c="red", linestyle='solid')
show_plot()
```

```python
data = pd.read_csv('data.csv')
price_date = data['Date']
price_close = data['Close']
```

```python
def show_plot():
  plt.gcf().autofmt_xdate()
  plt.title('Bitcoin Prices')
  plt.xlabel('Date')
  plt.ylabel('Closing Price')

  plt.tight_layout()

  plt.show()
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')
show_plot()
```
